```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('images/input.jpg')

hist=cv2.calcHist([image],[0],None,[256],[0,256])

plt.hist(image.ravel(),256,[0,256])
plt.show()

# we infer that the colors in this image are in a lower range 
```

```python
## plots for all of the colors
```

```python
d={0:'b',1:'r',2:'g'}


for i in d:
    hist=cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(hist,color=d[i])
plt.show()
    
```
